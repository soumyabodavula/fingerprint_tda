```python
import fingerprint
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.array([[[0, 0], [1, 1], [3, 2], [2, 3], [3, 3], [2, 2]]])
```

```python
from gtda.plotting import plot_point_cloud

plot_point_cloud(points[0])
```

```python
from gtda.homology import VietorisRipsPersistence

# Track connected components, loops, and voids
homology_dimensions = [0, 1, 2]

# Collapse edges to speed up H2 persistence calculation!
persistence = VietorisRipsPersistence(
    metric="euclidean",
    homology_dimensions=homology_dimensions,
    n_jobs=6,
    collapse_edges=True,
)

diagrams_basic = persistence.fit_transform(points)
```

```python
from gtda.plotting import plot_diagram

plot_diagram(diagrams_basic[0])
```

```python
def To2DArr(img):
    im = np.zeros((len(img), len(img[0])), dtype=float)

    for r in range(len(img)):
        for c in range(len(img[r])):
            im[r][c] = np.mean(img[r][c])/255
    
    return im
```

```python
img = cv2.imread("images/test3.png", cv2.IMREAD_ANYCOLOR)
im = To2DArr(img) # shape: (620, 424, 3)
```

```python
# Binarizing finger print image
from skimage.util import invert
original_img = cv2.imread("images/test3.png", cv2.IMREAD_ANYCOLOR)
BnW_image = fingerprint.BinarizeFingerprint(original_img, method='Pixel Blocks Optimized', blockSize=15)
BnW_image = invert(BnW_image)
```

```python
centerPoint = (10, 10)

theta = np.linspace( 0 , 2 * np.pi , 150 )
radius = 658
a = radius * np.cos( theta ) + centerPoint[0]
b = radius * np.sin( theta ) + centerPoint[1]

plt.imshow(BnW_image, cmap='gray')

plt.scatter(centerPoint[0], centerPoint[1], c='r') # looking for a center for radial filtration manually
#plt.plot(a,b, linewidth = 2, alpha = 0.5)
plt.show()
```

```python
points = []
pointsX = []
pointsY = []
pointsZ = []

BnW_image_cpy = np.array([[[0]*len(BnW_image[0])]*len(BnW_image)])


for x in range(len(BnW_image)):
    for y in range(len(BnW_image[x])):
        if BnW_image[x][y] != 0:
            #print(BnW_image[x][y])
            points.append([x,y,])

points = np.array([points])
```
